```python
#Importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

2020 CON DESVIACION ESTANDAR

```python
#Cargar archivo
data=pd.read_csv("Gastos y costos 2020_Sin_Nulos.csv")
```

```python
data.head()
```

```python
data20= data.drop(["Unnamed: 0","Columna1"], axis=1)
data20
```

```python
#Identificar valores nulos por columna .isnull()
valores_nulos=data20.isnull().sum()
valores_nulos
```

```python
data20.info()
```

```python
Cuantitativas=data20.iloc[:,8:15]
```

```python
Cualitativas=data20.iloc[:,[0,1,2,3,4,5,6,7,15,16,17]]
```

```python
#Realizamos diagrama de caja y bigote de cada columna del dataframe
fig=plt.figure(figsize=(20,8))
Cuantitativas.plot(kind="box",vert=False)
plt.title("Valores Atipicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Metdo aplicando desviacion estandar para encontrar los limites permitidos, encuentro los valores externos
y=Cuantitativas
Limite_Superior=y.mean()+3*y.std()
Limite_Inferior=y.mean()-3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliners se convierten en nulos en el dataframe
Cuantitativas1=Cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
Cuantitativas1
```

```python
#Identificar valores nulos por columna .isnull()
valores_nulos=Cuantitativas1.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos (nulos) con mean
#Realizamos una copia del data
Cuantitativas_limpio=Cuantitativas1.copy()
Cuantitativas_limpio=Cuantitativas_limpio.fillna(round(Cuantitativas1.mean(),1))
Cuantitativas_limpio
```

```python
#Identificar valores nulos por columna .isnull()
valores_nulos=Cuantitativas_limpio.isnull().sum()
valores_nulos
```

```python
#Unimos valores cualitativos
Valores_limpios_2020=pd.concat([Cuantitativas_limpio,Cualitativas],axis=1)
Valores_limpios_2020
```

```python
Valores_limpios_2020.to_csv("Valores_atipicos_2020.csv")
```

```python
Cualitativas1=data21.iloc[:,[0,1,2,3,4,5,6,7,15,16,17]]
```

```python
#Realizamos diagrama de caja y bigote de cada columna del dataframe
fig=plt.figure(figsize=(15,8))
data21.plot(kind="box",vert=False)
plt.title("Valores Atipicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Metodo aplicando cuartiles encuentro cuartiles 0.25 y 0.75
y=Cuantitativas1

percentile25=y.quantile(0.25) #q1
percentile75=y.quantile(0.75) #q3
iqr=percentile75-percentile25

Limite_Superior_2021= percentile75+1.5*iqr
Limite_Inferior_2021=percentile25-1.5*iqr
print("Limite superior permitido", Limite_Superior_2021)
print("Limite inferior permitido", Limite_Inferior_2021)
```

```python
#Obtenemos datos y los outliners se convierten en nulos en el dataframe
Cuantitativas1_1=Cuantitativas1[(y<=Limite_Superior_2021)&(y>=Limite_Inferior_2021)]
Cuantitativas1_1
```

```python
#Identificar valores nulos por columna .isnull()
valores_nulos=Cuantitativas1_1.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del df con "mean"
#Realizamos copia del df
Cuantitativas_2021_limpio=Cuantitativas1_1.copy()
Cuantitativas_2021_limpio=Cuantitativas_2021_limpio.fillna(round(Cuantitativas1_1.mean(),1))
Cuantitativas_2021_limpio
```

```python
#Identificar valores nulos por columna .isnull()
valores_nulos=Cuantitativas_2021_limpio.isnull().sum()
valores_nulos
```

```python
Valores_limpios_2021=pd.concat([Cuantitativas_2021_limpio,Cualitativas1],axis=1)
Valores_limpios_2021
```

```python
Valores_limpios_2021.to_csv("Valores_atipicos_2021.csv")
```
